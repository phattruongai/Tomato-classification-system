```python
a = [28,17.1,16.4]
b = [0.3,4.4,9.9]
c = [100-76,100-79,100-80.9]
d = [4.1,8.4,32]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(b,a)
plt.plot(b,a)
plt.scatter(d,c)
plt.plot(d,c)
```

```python
n1 = ['MobileNetV2','EfficientNet-B4','EfficientNet-B5']
n2 = ['Resnet50','Xception','ResNeXt-101']
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

plt.scatter(b,a)
plt.plot(b,a)
plt.scatter(d,c)
plt.plot(d,c)
plt.xlim(right = 40)
plt.xlabel('FLOPS(Bilion)')
plt.ylabel('Top-1 Accuracy(%)')
for i,xy in enumerate(zip(b, a)):                                       # <--
    ax.annotate(f'{n1[i]}', xy=xy, textcoords='data')
for i,xy in enumerate(zip(d, c)):                                       # <--
    ax.annotate(f'{n2[i]}', xy=xy, textcoords='data')
```
